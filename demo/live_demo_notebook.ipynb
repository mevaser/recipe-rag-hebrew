## 3. Run the Demo

Run the following cell to execute the existing demo script with the recommended Hebrew question. The cell is included for live use, but it is not executed automatically in this notebook file.

```python
import os
import subprocess
from pathlib import Path

PROJECT_ROOT = Path(r"C:\Users\mevas\Desktop\pythonProject\AI_Course\RAG")
os.chdir(PROJECT_ROOT)

cmd = [
    "python",
    "demo/run_demo_query.py",
    "--question",
    "מה כמות הקמח, המים והשמרים בבצק הפיצה?",
]

result = subprocess.run(
    cmd,
    capture_output=True,
    text=True,
    encoding="utf-8",
    errors="replace",
)

print(result.stdout)
if result.stderr:
    print("STDERR:")
    print(result.stderr)
```
